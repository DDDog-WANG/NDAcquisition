# Super Pixels

```python
img = VPAtreaded_img
felzen_segments = felzenszwalb(img)
quick_segments = quickshift(img)
slic_segments = slic(img,n_segments=50)
water_segments = watershed(sobel(rgb2gray(img)), markers=250) 
```

```python
plt.figure(figsize=(10, 10))

plt.subplot(2, 2, 1)
plt.title("Felzenszwalb")
plt.imshow(mark_boundaries(img,felzen_segments))

plt.subplot(2, 2, 2)
plt.title("quickshift")
plt.imshow(mark_boundaries(img,quick_segments))

plt.subplot(2, 2, 3)
plt.title("slic")
plt.imshow(mark_boundaries(img,slic_segments))

plt.subplot(2, 2, 4)
plt.title("watershed")
plt.imshow(mark_boundaries(img,water_segments))
```

```python
n_segments=100

plt.figure(figsize=(20,20))
slic_segments = slic(untreaded_img,n_segments=n_segments)
plt.subplot(1, 3, 1)
plt.axis("off")
plt.imshow(slic_segments)
plt.subplot(1, 3, 2)
plt.axis("off")
plt.title("untreaded_img")
plt.imshow(untreaded_img)
plt.subplot(1, 3, 3)
plt.axis("off")
plt.imshow(mark_boundaries(untreaded_img,slic_segments))

plt.figure(figsize=(20,20))
slic_segments = slic(VPAtreaded_img,n_segments=n_segments)
plt.subplot(1, 3, 1)
plt.axis("off")
plt.imshow(slic_segments)
plt.subplot(1, 3, 2)
plt.axis("off")
plt.title("VPAtreaded_img")
plt.imshow(VPAtreaded_img)
plt.subplot(1, 3, 3)
plt.axis("off")
plt.imshow(mark_boundaries(VPAtreaded_img,slic_segments))
```
