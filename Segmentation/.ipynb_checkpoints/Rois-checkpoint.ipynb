# Transform boxes and masks from 512 => 2560

## 1. box in 512 ==> 2560

```python
# box in 512 ==> 2560
box_2560=[]
for x in range(box_512.shape[0]):
    box_2560.append([box_512[x][0]*5, box_512[x][1]*5, (box_512[x][2]+1)*5-1, (box_512[x][3]+1)*5-1])
box_2560 = np.array(box_2560)

print("box_2560.shape：", box_2560.shape)
print("box_512[1] before：", box_512[1])      
print("box_512[1] after：", box_2560[1])
print("")
```

## 2. mask 512 * 512 ==> 2560 * 2560

```python
mask_2560 = []

for i in range(mask_512.shape[2]):
    mask = mask_512[:,:,i]
    img_mask = np.zeros((2560,2560))
    for x in range(512):
        for y in range(512):
            if mask[x][y] == True:
                for m in range(5*x,5*(x+1)):
                    for n in range(5*y,5*(y+1)):
                        img_mask[m][n]=1
    mask_2560.append(img_mask)
    
mask_2560 = np.array(mask_2560)
print("mask_2560.shape：", mask_2560.shape)
```

## 3.1 Using mask 

```python
img = cv2.imread("../../Datasets/211202NDAcquisition/NDAcquisition-01/NDAcquisition-01_XY212.ome.tif")
img = cv2.copyMakeBorder(img,128,128,128,128,cv2.BORDER_CONSTANT,value=(0,0,0))

i=10
pic = img[box_2560[i][0]:box_2560[i][2], box_2560[i][1]:box_2560[i][3]]
print("pic.shape",pic.shape)
plt.imshow(pic)
plt.show()

mask = mask_2560[i,:,:]
mask = mask[box_2560[i][0]:box_2560[i][2], box_2560[i][1]:box_2560[i][3]]
mask = cv2.merge([mask,mask,mask])
print("mask.shape",mask.shape)
plt.imshow(mask)
plt.show()


pic_mask=np.multiply(pic,mask)
pic_mask=pic_mask.astype(np.uint8)
print("pic_mask.shape",pic_mask.shape)
plt.imshow(pic_mask)
plt.show()
```

## 3.2 Using opencv to get ellipse from mask

```python
img = cv2.cvtColor(pic_mask,cv2.COLOR_BGR2GRAY)
ret, binary  = cv2.threshold(img,1,1,0)
contours,hierarchy = cv2.findContours(binary, 1, 2)

piccontours=cv2.drawContours(mask,contours,-1,(0,255,0),1) 
plt.imshow(piccontours)
plt.show()
```

```python
ellipse = cv2.fitEllipse(contours[0])
img = cv2.ellipse(mask,ellipse,(0,255,0),1)
plt.imshow(img)
plt.show()

img = cv2.ellipse(pic,ellipse,(0,255,0),1)
plt.imshow(img)
plt.show()
```

## 3.3 using opencv to get binary mask and merge to ROI

```python
img = cv2.cvtColor(pic,cv2.COLOR_BGR2GRAY)
ret, binary  = cv2.threshold(img,50,1,0)
mask_binary = cv2.merge([binary,binary,binary])
plt.imshow(binary)
plt.show()

pic_mask=np.multiply(pic,mask_binary)
pic_mask=pic_mask.astype(np.uint8)
plt.imshow(pic_mask)
plt.show()

plt.imshow(pic)
plt.show()
```

## 3.4 using opencv to get contours

```python
contours,hierarchy = cv2.findContours(binary, 1, 2)
piccontours=cv2.drawContours(pic,contours,-1,(0,255,0),1) 
plt.imshow(piccontours)
plt.show()
```

```python
cnt = contours[0]
# moments
M = cv2.moments(cnt)
# area
area = cv2.contourArea(cnt)
print(area)
# perimeter
perimeter = cv2.arcLength(cnt,True)
print(perimeter)
```

## 3.5 using opencv to draw minEnclosingCircle

```python
(x,y),radius = cv2.minEnclosingCircle(cnt)
center = (int(x),int(y))
radius = int(radius)
img = cv2.circle(pic,center,radius,(0,255,0),1)
plt.imshow(img)
plt.show()
```

## 3.6 using opencv to draw ellipse

```python
ellipse = cv2.fitEllipse(cnt)
img = cv2.ellipse(pic,ellipse,(0,255,0),1)
plt.imshow(img)
plt.show()
```
